```python
from sklearn.feature_extraction.text import TfidfTransformer, TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
documents = [
    "This is a list which contains sample documents",
    "keywords are important for keyword based search",
    "Document analysis involves extracting keywords",
    "keyword based search relies on sparse embedding",
]
```

```python
query = "keyword based search"
```

```python
import re
def preprocess_text(text:str):
    text = text.lower()
    #remove punctuation
    text = re.sub(r'[^\w\s]','',text) #Anything that is NOT:- letter - digit - underscore - whitespace ^ means not (return all words that has no special character)
    return text
```

```python
preprocessed_documents = [preprocess_text(doc) for doc in documents]
preprocessed_documents
```

```python
processed_query = preprocess_text(query)
processed_query
```

```python
#vector = TfidfTransformer() #expects 2D
# vector.fit_transform(preprocessed_documents) #TfidfTransformer expects a document-term count matrix, not raw text. Rows = documents and Columns = words and Values = word counts
```

```python
vector = TfidfVectorizer()
x = vector.fit_transform(preprocessed_documents) #TfidfTransformer expects a document-term count matrix, not raw text. Rows = documents and Columns = words and Values = word counts
```

```python
x
```

```python
print(f"Shape of the x vector for 1D documents : {x.shape} follows")
x.toarray()[0]  #4 rows and 22 columns
```

```python
query_embedding = vector.transform([processed_query])
query_embedding.shape
```

```python
similarities = cosine_similarity(x, query_embedding)
similarities
```

```python
similarities.shape  #4 rows 1 column
```

```python
ranked_indices = np.argsort(similarities, axis=0)[::-1].flatten()  # descending [::-1]  np.argsort(similarities, axis=0) ascending
print(f"Ranked indices are :{ranked_indices}")
ranked_documents = [preprocessed_documents[index] for index in ranked_indices]
```

```python
for i, doc in enumerate(ranked_documents):
    print(f"Rank {i+1}: {doc}")
```
